```python
#Import Dependencies
from pylab import *
import os
import pandas as pd
# import librosa
# import librosa.display
# import glob
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical
```

# Model

```python
df = pd.read_csv('../data/gender_data.csv')
df.head()
```

```python
df.drop(inplace=True, columns=["emotions"])
df.head()
```

```python
target = df["label"]
```

```python
data = df.drop("label", axis=1)
feature_names = data.columns
data.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)
```

```python
y_test
```

```python
from joblib import load

X_scaler = load('scaler.sav')
```

```python
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# One-hot encoding
y_train_categorical = to_categorical(y_train)
y_test_categorical = to_categorical(y_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()

model.add(Dense(units=256, activation='relu', input_dim=128))
model.add(Dense(units=256, activation='relu'))
model.add(Dense(units=y_train_categorical.shape[1], activation='softmax'))
```

```python
model.summary()
```

```python
# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Fit the model to the training data
model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=100,
    shuffle=True,
    verbose=2
)
```

```python
model_loss, model_accuracy = model.evaluate(
    X_test_scaled, y_test_categorical, verbose=2)
print(
    f"Gender Deep Learning - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
from sklearn.metrics import classification_report
predictions = model.predict(X_test_scaled)
print(classification_report(y_test_categorical, predictions,))
```

```python
import joblib
filename = 'deep_learning_gender_model.sav'
joblib.dump(model, filename)
```
